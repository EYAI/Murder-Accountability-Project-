### Perpetrator Profil

```python
data['Perpetrator Age'] = pd.to_numeric(data['Perpetrator Age'])
```

```python
data['Perpetrator Age'] = data['Perpetrator Age'].astype("int")
mask = (data['Perpetrator Age'] < 18)
young_Perpetrator =  pd.DataFrame(data.loc[mask], columns = ['Year']) 
count_years = young_Perpetrator.stack().value_counts()
homicides_young = count_years.sort_index(axis=0, ascending=False)
mask2 = (data['Perpetrator Age'] > 18)
adult_Perpetrator =  pd.DataFrame(data.loc[mask2], columns = ['Year']) 
count_years = adult_Perpetrator.stack().value_counts()
homicides_adult = count_years.sort_index(axis=0, ascending=False)
print(homicides_adult.plot(kind='barh', fontsize=10,  width=0.5,  figsize=(12, 10), title='Perpetrator above 18 years old'))
```

#### On peut remarquer ainsi que les tueurs deviennet de plus en plus jeunes d'une annee a une autre .

```python
homicides_adult.to_frame()
homicides_young.to_frame()
homicides = pd.DataFrame({'Adult': homicides_adult,'Young':homicides_young})
homicides.sort_index(inplace=True)
pos = list(range(len(homicides['Adult'])))
width = 0.15

# Plotting the bars
fig, ax = plt.subplots(figsize=(25,15))

# in position pos,
plt.bar(pos,
        #using homicides['Adult'] data,
        homicides['Adult'],
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#EE3224',
        # with label the first value in year
        label=homicides.index[0])


# Create a bar with young data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos],
        #using homicides['Young'] data,
        homicides['Young'],
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#F78F1E',
        # with label the second value in year
        label=homicides.index[1])



# Set the y axis label
ax.set_ylabel('Adult / Young')

# Set the chart's title
ax.set_title('Comparation between victims by age')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(homicides.index)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(homicides['Adult'] + homicides['Young'])] )

# Adding the legend and showing the plot
plt.legend(['Adult', 'Young'], loc='upper left')

plt.grid()
plt.show()
```

#### Ceci confirme nos dires mais aussi s'ajoute que le nombre de tueurs diminuent d'une annee a une autre .
